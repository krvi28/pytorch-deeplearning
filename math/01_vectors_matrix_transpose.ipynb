```python
import numpy as np
import torch
```

#### Using Numpy

```python
# create a vector
nv = np.array([ [1, 2, 3, 4] ])
print(nv), print(' ')

# transpose a vector using numpy
print(np.transpose(nv))
```

```python
# transpose a vector
print(nv.T), print(' ')

nvT = nv.T

# Transpoe a transpose
print(nvT.T)
```

```python
# repeat for a matrix
nM = np.array([ [1, 2, 3, 4], [5, 6, 7, 8]])

print(nM), print(' ')

# transpose the matrix
print(nM.T), print(' ')

# transpose the transpose
nMT = nM.T
print(nMT.T)
```

#### Using Torch

```python
# transpose a matrix
tM = torch.tensor([ [1, 2, 3, 4],
                    [5, 6, 7, 8]])

print(tM), print(' ')

tMT = tM.T
print(tMT), print(' ')

# transpose a transpose
print(tMT.T)
```

```python
## examine the data types
print(f"Variable nv is of the data type: {type(nv)}")
print(f"Variable nM is of the data type: {type(nM)}")
print(f"Variable tM is of the data type: {type(tM)}")
```
