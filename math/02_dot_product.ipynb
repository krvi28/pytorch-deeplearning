```python
# import libraries
import numpy as np
import torch
```

#### Using Numpy

```python
# Define the vectors
nv1 = np.array([1, 2, 3, 4])
nv2 = np.array([0, -1, 0, -1])

# Dot product using dot function
print(np.dot(nv1, nv2))

# Dot product using sum function
print(np.sum(nv1 * nv2))
```

#### Using PyTorch

```python
# Define the vectors
tv1 = torch.tensor([1, 2, 3, 4])
tv2 = torch.tensor([0, -1, 0, -1])

# Dot product using dot function
print(torch.dot(tv1, tv2))

# Dot product using sum function
print(torch.sum(tv1 * tv2))
```
